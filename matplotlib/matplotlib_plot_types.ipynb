### Matplotlib: Many plot types
- How to compare categories of data with bar graphs?
- Add error bars to graphs
- Use fill_between to display shaded error on line graphs
- Create stacked bar graphs for easier comparisons
- Create side-by-side bar graphs
- Create and format pie charts to compare proportional datasets 
- Analyze frequency data using histograms, including multiple histograms on the same plot
- Normalize histograms

```python
# Import Matplotlib library:
import matplotlib.pyplot as plt
```

```python
# An instance of RcParams for handling default matplotlib values:
plt.rcParams.update({'font.size' : 14})
```

```python
# Simple Bar Chart: plt.bar()
# To create simple bar charts to compare multiple categories of data.

products = ['Coffee', 'Sugar', 'Pasta', 'Fruits', 'Milk', 'Fish']
sales = [65, 1765, 1544, 345, 433, 1536]

plt.bar(range(len(sales)), sales)
plt.show()
```

```python
# Simple Bar Chart: plt.bar() + ax = plt.subplot()
# Display category of data!

products = ['Coffee', 'Sugar', 'Pasta', 'Fruits', 'Milk', 'Fish']
sales = [65, 1765, 1544, 345, 433, 1536]

#create our ax object - Important, we need to call ax object before plt.plot!
ax = plt.subplot()
ax.set_xticks(range(len(products)))
ax.set_xticklabels(products)

plt.bar(range(len(products)), sales)
plt.show()
```

```python
# Side-By-Side Bars:
# We can use a List Comprehensions like it :
# n = ?  # This is our second dataset (out of 2)
# t = ? # Number of datasets
# d = ? # Number of sets of bars
# w = ? # Width of each bar
# x_values = [t*element + w*n for element in range(d)]

sales1 = [2.4, 3.6, 2.9, 4.7, 6.3]
sales2 = [4.8, 6.6, 7.3, 8.5, 9.9]
months = ['January', 'February', 'March', 'April', 'June']
pos = list(range(len(sales1)))
width = 0.4

xshop1 = pos
xshop2 = [p + 0.4 for p in pos]

fig, ax = plt.subplots(figsize=(10,5))
plt.bar(xshop1, sales1, width, alpha=0.5, label = 'Shop1')
plt.bar(xshop2, sales2, width, alpha=0.5, label = 'Shop2')

ax.set_ylabel('Sales K$')
ax.set_title('Shop1 vs Shop2')

# Set the position of the x ticks
ax.set_xticks([p + 0.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(months)

plt.legend(['Shop1', 'Shop2'])
plt.grid(linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Stacked Bars Version num. 1: "A poor graph, without customization"
# Compare two sets of data while preserving knowledge of the total between them, 
# we can also stack the bars instead of putting them side by side. 
sales1 = [2.4, 3.6, 2.9, 4.7, 6.3]
sales2 = [4.8, 6.6, 7.3, 8.5, 9.9]

plt.bar(range(len(sales1)), sales1)
plt.bar(range(len(sales2)), sales2, bottom = sales1)
plt.legend(['Shop1', 'Shop2'])
plt.show()
```

```python
# Stacked Bars Version num. 2: "This time we use an axes object, and many adaptations - Check out these differences"
# It's more beautiful, isn't it!?

sales1 = [2.4, 3.6, 2.9, 4.7, 6.3]
sales2 = [4.8, 6.6, 7.3, 8.5, 9.9]
months = ['January', 'February', 'March', 'April', 'June']

fig, ax = plt.subplots(figsize=(10,5))

plt.bar(range(len(sales1)), sales1, color='#FFC122', alpha = 0.5, label = 'Shop1')
plt.bar(range(len(sales2)), sales2, color='#F98F1E', alpha = 0.5, label = 'Shop2')

ax.set_xticks(range(5))
ax.set_xticklabels(months)

ax.set_ylabel('Sales K$')
ax.set_title('Shop1 vs Shop2')

plt.legend(['Shop1', 'Shop2'])
plt.grid(linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Now, we can add "Error Bars": with keywords yerr ans capsize
# Quickly with last beautiful graph, we add these parameters like down:

sales1 = [2.4, 3.6, 2.9, 4.7, 6.3]
sales2 = [4.8, 6.6, 7.3, 8.5, 9.9]
months = ['January', 'February', 'March', 'April', 'June']

error = [0.9, 2, 1.2, 1, 0.7]


fig, ax = plt.subplots(figsize=(10,5))

plt.bar(range(len(sales1)), sales1, color='#FFC122', alpha = 0.5, label = 'Shop1')
plt.bar(range(len(sales2)), sales2, color='#F98F1E', alpha = 0.5, label = 'Shop2',  yerr = error, capsize = 4)

ax.set_xticks(range(5))
ax.set_xticklabels(months)

ax.set_ylabel('Sales K$')
ax.set_title('Shop1 vs Shop2')

plt.legend(['Shop1', 'Shop2'])
plt.grid(linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Fill Between: plt.fill_between to shade error on line graph
# This is the same way as last it, but it's adapted to a line
# Fill-between is used to let us see a possible range of upper 
# and lower bounds that values might fall between, due to possible error or uncertainty.
import random

x = range(1, 100, 10)
y = random.sample(range(1,100), 10)
plt.figure(figsize=(15, 5))

plt.plot(x, y)

# To be plus or minus 10% of each value, so create a list containing the lower bound 
# of the expected revenue for each month, and call it y_lower.
y_lower = [i - (i*0.10) for i in y]
y_upper = [i + (i*0.10) for i in y]
plt.fill_between(x, y_lower, y_upper, alpha=0.2)

plt.title('Example fill between Matplotlib')
plt.show()
```

```python
# Pie Chart: plt.pie()
# Display elements of a data set as proportions of a whole, we can use a pie chart.
payment_type = ["Card", "Cheque", "Cash", "Virement"]
payment_webshop = [568, 97, 256, 46]

fig = plt.figure(figsize=(15,5))

plt.pie(payment_webshop, autopct='%0.1f%%')
plt.axis('equal')
plt.legend(payment_type)
plt.show()

# An other possibility with labels keyword (instead of legend):
fig = plt.figure(figsize=(15,5))

plt.pie(payment_webshop, labels = payment_type, shadow = False, explode=(0, 0, 0.15, 0), autopct='%0.1f%%')
plt.axis('equal')
plt.legend(payment_type)
plt.show()
```

```python
# Histogram: plt.hist()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

### Preliminaries ###

# Set ipython's max row display
pd.set_option('display.max_row', 1000)

# Set iPython's max column width to 50
pd.set_option('display.max_columns', 50)

df = pd.read_csv('https://data.cms.gov/resource/ehrv-m9r6.csv')
df.head()
# from : https://data.cms.gov/Medicare-Inpatient/Inpatient-Prospective-Payment-System-IPPS-Provider/97k6-zzx3
```

```python
# Make plot with bins of fixed size
data1 = df.average_medicare_payments
'''
Average Medicare Payments: The average amount that Medicare pays to the provider for Medicare's
share of the MS-DRG. Average Medicare payment amounts include the MS-DRG amount,
teaching, disproportionate share, capital, and outlier payments for all cases. Medicare payments DO
NOT include beneficiary co-payments and deductible amounts nor any additional payments from third
parties for coordination of benefits. Note: In general, Medicare FFS claims with dates-of-service or
dates-of-discharge on or after April 1, 2013, incurred a 2 percent reduction in Medicare payment. This is
in response to mandatory across-the-board reductions in Federal spending, also known as
sequestration. 
'''

# Create bins of 500 each
bins = np.arange(data1.min(), data1.max(), 500) # fixed bin size

fig, ax = plt.subplots(figsize=(15,5))

# Plot a histogram of Average Medicare Payments:
plt.hist(data1, 
         bins=bins, 
         alpha=0.5, 
         color='red',
         label='Medicare Payments')

ax.set_xlabel('Average $')
ax.set_title('Average Medicare Payments')

plt.show()
```

```python
# Multiple Histograms!

data1 = df.average_medicare_payments
data2 = df.average_medicare_payments_2

# Create bins of 500 each
bins = np.arange(data1.min(), data1.max(), 500) # fixed bin size

fig, ax = plt.subplots(figsize=(15,5))

# Plot a histogram of Average Medicare Payments:
plt.hist(data1, 
         bins=bins, 
         alpha=0.5, 
         color='Purple',
         label='Medicare Payments 1')

# Plot a histogram of Average Medicare Payments 2:
plt.hist(data2, 
         bins=bins, 
         alpha=0.5,
         color='#887E43',
         label='Medicare Payments 2')

ax.set_xlabel('Average $')
ax.set_title('Average Medicare Payments')
plt.legend(loc='upper right')
plt.show()
```

```python
# Multiple Histograms with smart argument: normed = True
# The result is two histograms that are very difficult to compare:
'''
Average Covered Charges: The provider's average charge for services covered by Medicare for all
discharges in the MS-DRG. These will vary from hospital to hospital because of differences in hospital
charge structures.
'''
data1 = df.average_medicare_payments
data2 = df.average_covered_charges

# Create bins of 500 each
bins = np.arange(data1.min(), data1.max(), 500) # fixed bin size

fig, ax = plt.subplots(figsize=(15,5))

# Plot a histogram of Average Medicare Payments:
plt.hist(data1, 
         bins=bins, 
         alpha=0.5, 
         color='Purple',
         #normed = True,
         label='Medicare Payments')

# Plot a histogram of Average Covered Charges:
plt.hist(data2, 
         bins=bins,          
         color='#887E43',
         #normed = True,
         label='Covered Charges')

ax.set_xlabel('Average $')
ax.set_title('Average Medicare Payments / Covered Charges')
plt.legend(loc='upper right')
plt.show()
```

```python
# Multiple Histograms with smart argument: normed = True
# To solve this, we can normalize our histograms using normed=True. 
# This command divides the height of each column by a constant such that 
# the total shaded area of the histogram sums to 1.

data1 = df.average_medicare_payments
data2 = df.average_covered_charges

# Create bins of 500 each
bins = np.arange(data1.min(), data1.max(), 500) # fixed bin size

fig, ax = plt.subplots(figsize=(15,5))

# Plot a histogram of Average Medicare Payments:
plt.hist(data1, 
         bins=bins, 
         alpha=0.5, 
         color='Purple',
         normed = True,
         label='Medicare Payments')

# Plot a histogram of Average Covered Charges:
plt.hist(data2, 
         bins=bins, 
         alpha=0.5,
         color='#887E43',
         normed = True,
         label='Covered Charges')

ax.set_xlabel('Average $')
ax.set_title('Average Medicare Payments / Covered Charges')
plt.legend(loc='upper right')
plt.show()
```
